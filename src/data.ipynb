# Visualise the Jet data from the CMS detector

```python
import numpy as np
import pandas as pd
import glob
import h5py
import matplotlib.pyplot as plt
from sklearn import preprocessing
%matplotlib inline
```

## A single event is shown with each event having 16 particles and 3 features for each particle

```python
h5file = h5py.File("../../data/bkg_3mln.h5", "r")

data = h5file["jetConstituentsList"][()]
features = h5file["particleFeatureNames"][()]

h5file.close()

data[0]
```

## Plot data


```python
def makePlot_p(feature_index, input_data, input_featurenames):
    plt.subplots()
    my_data = input_data[:,:,feature_index]
    plt.hist(my_data[:,feature_index].flatten(), 50, density=True, histtype='step', fill=False, linewidth=1.5)
    plt.yscale('log', nonposy='clip')     
    plt.xlabel(str(input_featurenames[feature_index], "utf-8"), fontsize=15)
    plt.ylabel('Prob. Density (a.u.)', fontsize=15)
    plt.show()
```

```python
for i in range(len(features)):
    makePlot_p(i, data, features)
```
